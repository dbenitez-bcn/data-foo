```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)

def f(x):
    return x**2

def g(x):
    return x + np.log2(x)

y = g(f(x))
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
x = np.linspace(-10, 10, 100)

def f(x):
    tmp = np.zeros(100)
    for idx, x in enumerate(x):
        if x > 0:
            tmp[idx] = x**7
        else:
            tmp[idx] = -x**7
    return tmp
y = f(x)
fig, axes = plt.subplots(1,2)
axes[0].plot(x,x**8)
axes[0].grid()
axes[1].plot(x,y)
# axes[1].plot(x,x**8)
axes[1].grid()
plt.show()
```

```python
x = np.linspace(-10,10, 100)

def f(x):
    return 2**x - x**4 

y = f(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(start=-5, stop=6)

def f(x):
    return x**2
y = f(x+2)
plt.plot(x, y)
plt.grid()
```
