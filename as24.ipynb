```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.DataFrame({
    "Total appointmnets": [5455, 7676-5455, 7676],
    "Total dealers affected": [166, 324-166, 324],
    "Total appointments - April": [1270, 1333-1270, 1333],
    "Total dealers affected - April": [154, 166-154, 166]
}, index=["Billed", "Not billed", 'Total'])
```

```python
labels = 'Billed', 'Lost'
explode = (0, 0.1)

fig1, ax = plt.subplots(1,2, figsize=(10, 5))
# Lost appointments
ax[0].pie([5455, 7676-5455], explode=(0, 0.1), labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax[0].set_title("Appointmnet billed (Nov-Apr)")
ax[0].axis('equal')

# Lost appointments
ax[1].pie([1270, 1333-1270], explode=(0, 0.3), labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax[1].set_title("Appointmnet billed (April)")
ax[1].axis('equal')

plt.savefig("appointments.png")

fig, ax = plt.subplots()
labels = ['Nov-Apr', 'April']
men_means = [5455, 1270]
women_means = [7676-5455, 1333-1270]
width = 0.75       # the width of the bars: can also be len(x) sequence
ax.bar(labels, men_means, width, label='Billed')
ax.bar(labels, women_means, width, bottom=men_means,
       label='Not billed')
ax.set_title('Dealers affected (Approximated)')
ax.legend()

plt.savefig("dealers.png")
plt.show()
```
